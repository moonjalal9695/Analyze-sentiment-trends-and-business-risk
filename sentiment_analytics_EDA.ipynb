```python
from textblob import TextBlob

def get_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity
    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"

df['sentiment_tag'] = df['Sentiment'].apply(get_sentiment)

```

```python
df.head()
```

```python
df['sentiment_score'] = df['sentiment_tag'].map({
    'Negative': -1,
    'Neutral': 0,
    'Positive': 1
})
```
